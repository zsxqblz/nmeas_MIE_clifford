```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.size': 20})
plt.rcParams.update({'lines.linewidth': 2})
```

### BW Nmeas Depth

```python
file_name = "../data/240114/240114_1_SAB_"
n_meas_l = np.loadtxt(file_name+'scanx.csv',skiprows=1)
depth_l = np.loadtxt(file_name+'scany.csv',skiprows=1)
raw = np.loadtxt(file_name+'data.csv',skiprows=1,delimiter=',')
SAB_ave_l = raw[:,0].reshape((len(depth_l),len(n_meas_l)))
SAB_std_l = raw[:,1].reshape((len(depth_l),len(n_meas_l)))

file_name = "../data/240114/240114_1_SBC_"
raw = np.loadtxt(file_name+'data.csv',skiprows=1,delimiter=',')
SBC_ave_l = raw[:,0].reshape((len(depth_l),len(n_meas_l)))
SBC_std_l = raw[:,1].reshape((len(depth_l),len(n_meas_l)))

file_name = "../data/240114/240114_1_SB_"
raw = np.loadtxt(file_name+'data.csv',skiprows=1,delimiter=',')
SB_ave_l = raw[:,0].reshape((len(depth_l),len(n_meas_l)))
SB_std_l = raw[:,1].reshape((len(depth_l),len(n_meas_l)))

file_name = "../data/240114/240114_1_SABC_"
raw = np.loadtxt(file_name+'data.csv',skiprows=1,delimiter=',')
SABC_ave_l = raw[:,0].reshape((len(depth_l),len(n_meas_l)))
SABC_std_l = raw[:,1].reshape((len(depth_l),len(n_meas_l)))

cmi_ave_l = SAB_ave_l+SBC_ave_l-SB_ave_l-SABC_ave_l

# plt.imshow(cmi_ave_l,extent=[n_meas_l[0],n_meas_l[-1],depth_l[-1],depth_l[0]], aspect='auto')
# plt.xlabel("# measurements")
# plt.ylabel("circuit depth")

plt.rcParams["figure.figsize"] = (15,10)
plt.subplot(2,3,1)
plt.imshow(SAB_ave_l,extent=[n_meas_l[0],n_meas_l[-1],depth_l[-1],depth_l[0]], aspect='auto')
plt.xlabel("# measurements")
plt.ylabel("circuit depth")
plt.subplot(2,3,2)
plt.imshow(SBC_ave_l,extent=[n_meas_l[0],n_meas_l[-1],depth_l[-1],depth_l[0]], aspect='auto')
plt.subplot(2,3,3)
plt.imshow(SB_ave_l,extent=[n_meas_l[0],n_meas_l[-1],depth_l[-1],depth_l[0]], aspect='auto')
plt.subplot(2,3,4)
plt.imshow(SABC_ave_l,extent=[n_meas_l[0],n_meas_l[-1],depth_l[-1],depth_l[0]], aspect='auto')
plt.subplot(2,3,5)
plt.imshow(cmi_ave_l,extent=[n_meas_l[0],n_meas_l[-1],depth_l[-1],depth_l[0]], aspect='auto')
plt.xlabel("# measurements")
plt.ylabel("circuit depth")
plt.tight_layout()
```
