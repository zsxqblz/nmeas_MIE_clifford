### 2D Noisy BW

|C|=1 measure C dx=10, noise=0.01+0.025*(i), depth=5, dy=2:20:2, nsim=20000*100, 100+1,2,3,4

|C|=1 measure C dx=14, noise=0.01+0.025*(i), depth=5, dy=2:20:2, nsim=10000*100, 200+1,2,3,4

|C|=1 measure C dx=10, noise=0.01+0.05*(i), depth=10, dy=2:20:2, nsim=10000*100, 400+1,2,3,4

```python
file_name = "../data/250322/250322_n104"
raw = np.load(file_name+"_pp.npz")
dy_l = raw["dy_l"]
dist_ave_l = raw["dist_ave_l"]
dist_std_l = raw["dist_std_l"]


plt.figure(figsize=(6,5))
# plt.plot(dy_l,np.sqrt(dist_ave_l),'.-')
plt.errorbar(dy_l,np.sqrt(dist_ave_l),yerr=np.sqrt(dist_std_l))
plt.hlines(1,xmin=dy_l[0],xmax=dy_l[-1])
plt.yscale('log')

print((np.log2(np.sqrt(dist_ave_l))[0] - np.log2(np.sqrt(dist_ave_l))[-1])/18)

print(dist_ave_l,dist_std_l)
```
