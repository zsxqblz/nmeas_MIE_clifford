```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.size': 20})
plt.rcParams.update({'lines.linewidth': 2})
```

### BW Nmeas Depth

```python
file_name = "../data/230414/230414_1_"

n_meas_l = np.loadtxt(file_name+'scanx.csv',skiprows=1)
depth_l = np.loadtxt(file_name+'scany.csv',skiprows=1)
raw = np.loadtxt(file_name+'data.csv',skiprows=1,delimiter=',')
cmi_ave_l = raw[:,0].reshape((len(depth_l),len(n_meas_l)))


plt.rcParams["figure.figsize"] = (6,5)
plt.imshow(cmi_ave_l,extent=[n_meas_l[0],n_meas_l[-1],depth_l[-1],depth_l[0]], aspect='auto')
# plt.xlim([50,100])
plt.xlabel("# measurements")
plt.ylabel("circuit depth")
plt.tight_layout()
```

```python
file_name = "../data/230414/230414_1_"

n_meas_l = np.loadtxt(file_name+'scanx.csv',skiprows=1)
depth_l = np.loadtxt(file_name+'scany.csv',skiprows=1)
raw = np.loadtxt(file_name+'data.csv',skiprows=1,delimiter=',')
cmi_ave_l = raw[:,0].reshape((len(depth_l),len(n_meas_l)))

for i in range(0,len(depth_l),1):
    plt.plot(n_meas_l,cmi_ave_l[i,:])
# plt.yscale('log')

plt.rcParams["figure.figsize"] = (6,5)
# plt.xlim([50,100])
plt.yscale('log')
plt.xlabel("# measurements")
plt.ylabel("MIE")
plt.tight_layout()
```
