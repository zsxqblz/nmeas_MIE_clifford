```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.size': 20})
plt.rcParams.update({'lines.linewidth': 2})
```

### RandClif

```python
plt.rcParams["figure.figsize"] = (6,5)
for i in range(1,6):
    raw = np.loadtxt("data/230317/230317_%d"%(i),delimiter=',',skiprows=1)
    plt.plot(raw[:,0],raw[:,1]+1e-3)

plt.xlim([0,20])
# plt.ylim([1e-3,3e1])
plt.xlabel("number of measurements")
plt.ylabel("MIE")
# plt.yscale('log')
plt.tight_layout()
```
