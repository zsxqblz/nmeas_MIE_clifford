```python
file_name = "../data/250319/250319_300"
raw = np.loadtxt(file_name,skiprows=1,delimiter=',')
dy_l = raw[:,0]
dist_ave_l = raw[:,1]
dist_std_l = raw[:,2]

plt.figure(figsize=(6,5))
plt.plot(dy_l,2**(10*5)*np.sqrt(dist_ave_l),'.-')
plt.errorbar(dy_l,2**(10*5)*np.sqrt(dist_ave_l),yerr=2**(10*5)*np.sqrt(dist_std_l))
plt.hlines(1,xmin=dy_l[0],xmax=dy_l[-1])
plt.yscale('log')

print((np.log2(np.sqrt(dist_ave_l))[0] - np.log2(np.sqrt(dist_ave_l))[-1])/18)

print(dist_ave_l,dist_std_l)
```
