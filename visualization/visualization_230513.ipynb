```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.size': 20})
plt.rcParams.update({'lines.linewidth': 2})
```

### BW Nmeas Depth
no choi map 5 by 5

```python
file_name = "../data/230513/230513_1_SA_"
n_meas_l = np.loadtxt(file_name+'scanx.csv',skiprows=1)
depth_l = np.loadtxt(file_name+'scany.csv',skiprows=1)
raw = np.loadtxt(file_name+'data.csv',skiprows=1,delimiter=',')
SA_ave_l = raw[:,0].reshape((len(depth_l),len(n_meas_l)))
SA_std_l = raw[:,1].reshape((len(depth_l),len(n_meas_l)))

file_name = "../data/230513/230513_1_SC_"
raw = np.loadtxt(file_name+'data.csv',skiprows=1,delimiter=',')
SC_ave_l = raw[:,0].reshape((len(depth_l),len(n_meas_l)))
SC_std_l = raw[:,1].reshape((len(depth_l),len(n_meas_l)))

file_name = "../data/230513/230513_1_SAC_"
raw = np.loadtxt(file_name+'data.csv',skiprows=1,delimiter=',')
SAC_ave_l = raw[:,0].reshape((len(depth_l),len(n_meas_l)))
SAC_std_l = raw[:,1].reshape((len(depth_l),len(n_meas_l)))

cmi_ave_l = SA_ave_l+SC_ave_l-SAC_ave_l

plt.rcParams["figure.figsize"] = (15,5)
plt.subplot(1,3,1)
plt.imshow(SA_ave_l,extent=[n_meas_l[0],n_meas_l[-1],depth_l[-1],depth_l[0]], aspect='auto')
plt.xlabel("# measurements")
plt.ylabel("circuit depth")
plt.subplot(1,3,2)
plt.imshow(SAC_ave_l,extent=[n_meas_l[0],n_meas_l[-1],depth_l[-1],depth_l[0]], aspect='auto')
plt.subplot(1,3,3)
plt.imshow(SA_ave_l+SC_ave_l-SAC_ave_l,extent=[n_meas_l[0],n_meas_l[-1],depth_l[-1],depth_l[0]], aspect='auto')
plt.xlabel("# measurements")
plt.ylabel("circuit depth")
plt.tight_layout()
```

```python
plt.rcParams["figure.figsize"] = (15,5)
plt.subplot(1,3,1)
for i in range(0,len(depth_l),1):
    plt.plot(n_meas_l,SA_ave_l[i,:])
# plt.yscale('log')
plt.xlabel("# measurements")
plt.ylabel("SA")
plt.subplot(1,3,2)
for i in range(0,len(depth_l),1):
    plt.plot(n_meas_l,SAC_ave_l[i,:])
# plt.yscale('log')
plt.xlabel("# measurements")
plt.ylabel("SAC")
plt.subplot(1,3,3)
for i in range(0,len(depth_l),1):
    plt.plot(n_meas_l,cmi_ave_l[i,:])
# plt.yscale('log')
plt.xlabel("# measurements")
plt.ylabel("CMI")
plt.tight_layout()
```

```python
plt.rcParams["figure.figsize"] = (15,5)
plt.subplot(1,3,1)
for i in range(0,len(n_meas_l),1):
    plt.plot(depth_l,SA_ave_l[:,i])
# plt.yscale('log')
plt.xlabel("depth")
plt.ylabel("SA")
plt.subplot(1,3,2)
for i in range(0,len(n_meas_l),1):
    plt.plot(depth_l,SAC_ave_l[:,i])
# plt.yscale('log')
plt.xlabel("depth")
plt.ylabel("SAC")
plt.subplot(1,3,3)
for i in range(0,len(n_meas_l),1):
    plt.plot(depth_l,cmi_ave_l[:,i])
# plt.yscale('log')
plt.xlabel("depth")
plt.ylabel("CMI")
plt.tight_layout()
```
