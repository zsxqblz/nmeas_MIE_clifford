```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.optimize import curve_fit
from scipy.stats import linregress
```

```python
def findCritPt(depth_l,cmi_ave):
    mid_pt = np.max(cmi_ave)/2
    mid_idx = np.argmin(np.abs(cmi_ave-mid_pt))
    # print(mid_idx)
    # return depth_l[mid_idx] - cmi_ave[mid_idx]*(depth_l[mid_idx+5]-depth_l[mid_idx-5])/(cmi_ave[mid_idx+5]-cmi_ave[mid_idx-5])
    # return depth_l[mid_idx]
    fit = linregress(depth_l[mid_idx-5:mid_idx+5],cmi_ave[mid_idx-5:mid_idx+5])
    return depth_l[mid_idx] - cmi_ave[mid_idx]/fit.slope
```

```python
N_l = [100,200,300,400,800,1600]
critPt_l = np.zeros(len(N_l))

for idx, i in enumerate([0,5,10,15,35,75]):
    file_name = "../data/240203/240203_n%d_"%(i)
    n_meas_l = np.loadtxt(file_name+'0_SAB_scanx.csv',skiprows=1)
    depth_l = np.loadtxt(file_name+'0_SAB_scany.csv',skiprows=1)

    raw = np.load(file_name+'SAB_pp.npz')
    n_meas_l = raw["n_meas_l"]
    depth_l = raw["depth_l"]
    SAB_ave_l = raw["data_ave_l"]
    SAB_std_l = raw["data_std_l"]

    raw = np.load(file_name+'SBC_pp.npz')
    SBC_ave_l = raw["data_ave_l"]
    SBC_std_l = raw["data_std_l"]

    raw = np.load(file_name+'SB_pp.npz')
    SB_ave_l = raw["data_ave_l"]
    SB_std_l = raw["data_std_l"]

    raw = np.load(file_name+'SABC_pp.npz')
    SABC_ave_l = raw["data_ave_l"]
    SABC_std_l = raw["data_std_l"]

    cmi_ave_l = SAB_ave_l+SBC_ave_l-SB_ave_l-SABC_ave_l

    center_idx = np.argmax(cmi_ave_l[-1,:])
    max_cmi = cmi_ave_l[:,center_idx]
    critPt_l[idx] = findCritPt(depth_l,max_cmi)

def fit_power(x,A,p):
    return A*(x**p)

popt, pcov = curve_fit(fit_power,N_l,critPt_l,p0=(2,0.5))


plt.rcParams["figure.figsize"] = (6,5)
plt.plot(N_l,critPt_l,"o")
plt.plot(N_l,fit_power(N_l,*popt),"--")
plt.xscale('log')
plt.yscale('log')
plt.xlim([80,2000])
plt.ylim([20,80])
plt.yticks()
plt.yticks([20,30,40,60,80],[20,30,40,60,80])
plt.xlabel('$N_B$')
plt.ylabel('critical depth')
plt.tight_layout()
```
